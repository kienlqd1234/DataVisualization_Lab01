### Sự phân bố giá trị của mỗi cột

#### Với mỗi cột có kiểu dữ liệu dạng numeric, các giá trị được phân bố như thế nào?

Với các cột có kiểu dữ liệu số, ta tính:

- Tỉ lệ % (từ 0 đến 100) các giá trị thiếu
- Giá trị min
- Giá trị lower quartile (phân vị 25)
- Giá trị median (phân vị 50)
- Giá trị upper quartile (phân vị 75)
- Giá trị max

Nhóm sẽ lưu kết quả vào DataFrame `num_col_info_df`, trong đó:
- Tên các cột là tên các cột số trong `raw_df`
- Tên các hàng: “missing_ratio”, “min”, “low_quartile”, “median”, “upper_quartile”, “max”

Để dễ quan sát, nhóm làm tròn tất cả các giá trị đến 1 chữ số thập phân bằng phương thức `.round(1)`.

```python
num_col_info_df = None

# Define missing ratio
def missing_ratio(series):
    percentage = series.isna().sum() / len(series)
    return percentage.round(1)

# Define lower quartile
def lower_quartile(series):
    return series.quantile(0.25).round(1)

# Define upper quartile
def upper_quartile(series):
    return series.quantile(0.75).round(1)

#Define median
def median(series):
    return series.quantile(0.75).round(1)

num_col_info_df = df.select_dtypes(include=['int64', 'float64'])

num_col_info_df = num_col_info_df.agg([missing_ratio, "min", lower_quartile, median, upper_quartile, "max"])

num_col_info_df
```

**Nhận xét**: Các cột có giá trị thiếu là: `GroundService`, `Food&Beverage`, `InflightEntertainment`, `Wifi&Connectivity`. Trong đó `Wifi&Connectivity` có tỉ lệ thiếu khá cao (80%) trong khi các cột còn lại có tỉ lệ thiếu khá thấp, trong mức chấp nhận được.

Sau khi xác định các giá trị thống kê cơ bản mô tả dữ liệu, chúng ta cần xác định thêm các thuộc tính có nhiều giá trị bị thiếu. Các thuộc tính như vậy không hữu ích cho giai đoạn phân tích và cần được xóa khỏi tập dữ liệu.

Tùy thuộc vào quy ước của mỗi người, ngưỡng "lớn" có thể được xác định. Thông thường, nếu tỷ lệ phần trăm giá trị bị thiếu lớn hơn 75%, cột đó sẽ bị loại khỏi dữ liệu.

```python
def drop_missing_features(df: pd.DataFrame, missing_lst: dict = dict(num_col_info_df.iloc[0]), threshold: float = 75.0) -> pd.DataFrame:
    """Drop missing features from Pandas dataframe base on given threshold.

    Args:
        df (pd.DataFrame): Input dataframe for processing.
        missing_lst (dict, optional): Dict missing value rate which calculated from previous stage. Defaults to dict(num_col_info_df.iloc[0]).
        threshold (float, optional): User threshold for dropping. Defaults to 75.0.

    Returns:
        pd.DataFrame: Updated dataframe
    """
    # YOUR CODE HERE
    for key, value in missing_lst.items():
        if value > (threshold/100):
            del df[key]
    return df
```

```python
df = drop_missing_features(df)
```

```python
df.head()
```

Sau khi xóa các thuộc tính có giá trị bị thiếu lớn, bộ dữ liệu của chúng ta vẫn còn các giá trị bị thiếu. Vì vậy, chúng ta cần điền những giá trị còn thiếu này để có thể sử dụng trong phân tích. Trong trường hợp này, vì đánh giá là số nguyên từ 1-5 nên nhóm sẽ chọn điền vào các giá trị thiếu bằng trung vị (median)

```python
def filling_missing_value(df: pd.DataFrame) -> pd.DataFrame:
    """Filling missing features from Pandas dataframe
    
    The first, you need to replace all None values with NaN, fillna only works on nans.
    
    After that, you replace all NaN values with the median of the column values.

    Args:
        df (pd.DataFrame): Input dataframe for processing.

    Returns:
        pd.DataFrame: Updated dataframe
    """
    # YOUR CODE HERE
    # Replace all None values with NaN
    df = df.replace('None', np.nan)
    
    # Replace all NaN values with the mean of the column values.
    for keys in df.keys():
        if df[keys].dtypes != 'float64':
            continue
        df[keys] = df[keys].fillna(df[keys].median())
    return df
    
```

```python
df = filling_missing_value(df)
```

#### Với mỗi cột có kiểu dữ liệu dạng categorical, các giá trị được phân bố như thế nào?

Với các cột có kiểu dữ liệu dạng `categorical`, ta cần tính:

- Tỉ lệ % (từ 0 đến 100) các giá trị thiếu
- Số lượng các giá trị khác nhau (không xét giá trị thiếu)
- Tỉ lệ % (từ 0 đến 100) của mỗi giá trị được sort theo tỉ lệ % giảm dần (không xét giá trị thiếu)

Chúng ra sẽ lưu kết quả vào DataFrame `cat_col_info_df`, trong đó:
- Tên các cột là tên các cột không phải số trong `raw_df`
- Tên các dòng là: “missing_ratio”, “num_values”, “value_ratios”

Để tiện quan sát, chúng ta làm tròn tất cả các giá trị đến 1 chữ số thập phân bằng phương thức `.round(1)`.

```python
pd.set_option('display.max_colwidth', 100) # For clearly
pd.set_option('display.max_columns', None) # For clearly
```

```python
cat_col_info_df = df.select_dtypes(include='object')

# Define num values
def num_values(series):
    return len(series.dropna().unique())

# Define value ratios
def value_ratios(series):
    value_counts = series.dropna().value_counts()
    percentages = (value_counts / len(series))*100
    percentages = percentages.round(1)
    sorted_percentages = percentages.sort_values(ascending=False)
    return sorted_percentages.to_dict()
cat_col_info_df = cat_col_info_df.agg([missing_ratio, num_values, value_ratios])
cat_col_info_df
```

**Nhận xét**: Các cột có giá trị thiếu là: `TypeOfTraveller	`, `Route`, `Aircraft`. Tỉ lệ thiếu của cột `Aircraft` khá cao nhưng vẫn chấp nhận được. Với các cột bị thiếu này, nhóm quyết định sẽ không điền vào các giá trị thiếu bởi đây là các thông tin không quá quan trọng và khả năng tiếp cận với thông tin chính xác cho những cột này là rất thấp. 

### Dữ liệu được thu thập có hợp lệ không?

Chúng ta có một thắc mắc liệu rằng số liệu thu thập được có hợp lý hay không? Ví dụ như các đánh giá (trừ OverallRating) có là số nguyên từ 1-5?

```python
for column in df.select_dtypes(include='number'):
    # Sử dụng all() để kiểm tra xem tất cả các giá trị đều nằm trong khoảng từ 1 đến 5 hay không
    if column == 'OverallRating':
        continue
    is_within_range = df[column].between(1, 5, inclusive=True).all()
    print(f"Cột {column}: {'Các số liệu nằm trong khoảng từ 1 đến 5' if is_within_range else 'Có số liệu nằm ngoài khoảng từ 1 đến 5'}")
```
